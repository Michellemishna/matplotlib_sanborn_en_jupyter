```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

Subplot, los parametros que les estoy pasando son [numero de filas],[numero de columnas][index que voy a graficar], en este caso grafica primero 'x' y despues 'y'

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.plot(y,x, 'b:')
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```
