```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

 la siguiente sentencia: axes = fig.add_axes([0.1,0.1,0.8,0.9])

Esto es lo que pasa (leyendo de izquierda a derecha)

El primer valor 0.1, significa que la figura comienza al 10% desde el lado izquierdo, en sentido horizontal (eje x) del lienzo

El segundo valor 0.1, significa que la figura comienza al 10% desde la parte inferior, en sentido vertical (eje y) del lienzo

El tercer valor 0.8, establece el ancho de la imagen que se extiende hasta el 80% del lienzo.

El cuarto valor 0.9 establece la altura de la imagen que se extiende hasta el 90% del lienzo.

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.3,0.3])
axes.plot(x,y,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.9,0.9,0.3,0.3])
axes.plot(x,y,'b')
```

con el orden de estas instrucciones le estoy diciendo que haga el axes2.plot dentro de axes.plot, son 2 objetos distintos dentro de el mismo lienzo el segundo sobrepuesto al primero.

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.6,0.4,0.3])
axes.plot(x,y,'b')
axes2.plot(y,x,'r')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.6,0.4,0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'r')
axes.set_facecolor('gray')
```
