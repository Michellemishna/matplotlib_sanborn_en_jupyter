```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

Este esta apegado al metodo orientado a objetos:
con figsize cambia el tamam√±o del lienzo,
con label le indicas que agregre una etiqueta y el "$" le da una notacion matematica, como cursiva
con .set_title() agregas un titulo al grafico
con .set_[x o y]label() agregas un titulo al eje
con.legend() le dices que te muestre las etiquetas en el grafico


```python
fig, axes = plt.subplots(1,2, figsize=(6,3))
axes[0].plot(x,y, label='$sin(x)$')
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

este es el metodo con pyplot basico

```python
plt.plot(x,y, label= 'sin(x)')
plt.title('Este es un titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc= 'lower left', bbox_to_anchor = (1,0.9))
```

ejemplo usando programacion y matplotlib orientado a objetos

```python
y = np.zeros(100)
x = np.linspace(0,20,100)
n = np.linspace(1,9,9, endpoint=True)
harmony = []
for i, element in enumerate(n):
    y = np.sin(x * element)
    harmony.append(y)
    
harmony = np.expand_dims(np.array(harmony), axis= 0)
harmony = harmony.reshape (3,3,100)

nrow = 3
ncol = 3
fig, axes = plt.subplots(nrow, ncol, figsize=(10,10))
for irow in range(nrow):
    for icol in range(ncol):
        axes[irow,icol].plot(x, harmony[irow, icol])
        axes[irow, icol].set_title('Harmony of $sin(x)$')
fig.tight_layout()
```
