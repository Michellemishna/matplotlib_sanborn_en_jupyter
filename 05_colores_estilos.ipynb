```python
import matplotlib.pyplot as plt
import numpy as np
```

con plt.style.available puedes ver los estilos ya predefinidos que tiene matplotlib

```python
x = np.linspace(0,5,11)
print(plt.style.available)
```

```python
plt.style.use('bmh')
```

# Matlab

```python
fig, ax = plt.subplots(figsize = (3,3))
ax.plot(x,x+1, 'r--')
ax.plot(x,x+2, 'b->')
ax.plot(x,x+3, 'y.-')
ax.plot(x,x+4, 'go:')
plt.show()
```

# Pyplot

acepta colores en HEX
y alpha sirve para agregar niveles de transparencia
linewidth le da el grozor a la linea 
linestyle es el estilo que tendra ma linea 
marker sera el estilo que tendran los puntos de union de nuestro grafico
markersize determina el tamaño que tendra marker
markerfacecolor cambia el color del marker

```python
fig, ax = plt.subplots(figsize = (3,3))
ax.plot(x,x+1, color = 'green', alpha = 0.4, linestyle= '-', marker = 'o', markersize = 9, markerfacecolor = '#34495E')
ax.plot(x,x+2, color = 'blue', linewidth=3, linestyle= '--', marker = 's', markersize = 5, markerfacecolor = '#16A085')
ax.plot(x,x+3, color = '#9B59B6', linewidth=1, linestyle= 'dashed', marker = 'p', markersize = 7, markerfacecolor = 'Salmon')
ax.plot(x,x+4, color = '#E59866', linewidth=2, linestyle= ':' , marker = '*', markersize = 3, markerfacecolor = 'c')
plt.show()
```
