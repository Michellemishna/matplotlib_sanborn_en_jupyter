```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['India', 'Japon', 'MÃ©xico', 'Colombia', 'Germany']
population = [1000, 800, 900, 1000, 300]
```

se puede determinar el color de cada columna agregado los colores que queremos tener en una lista, si queremos solo 2 colores estos los ira alternando
xticks sirve para que me de un arreglo con las posiciones que tengo, una vez las enlista le paso los nombres que quiero que tenga
rotation sirve para cambiar la posicion del titulo

```python
plt.bar(country, population, width=0.4, color = ['c', 'r', 'g', 'y', 'b'], label='population')
plt.xticks(np.arange(5), ('INDIA', 'JAPON','MEXICO','COLOMBIA', 'ALEMANIA'), rotation= 45)
plt.title('Population Country')
plt.ylabel('Population')
plt.legend(loc ='upper right')
plt.show()
```

barth cambia de vertical a horizontal los graficos

```python
plt.barh(country,population)
plt.show()
```

```python
print(plt.style.available)
plt.style.use('seaborn-v0_8')
fig, axes = plt.subplots()
bars = axes.barh(country,population, color=['#5c58b6','#b457ce','#5994ce','#3a4e93','#5DADE2'], label= 'Population country')
axes.bar_label(bars)
axes.set_title('Population Country')
axes.set_xlabel('Population')
axes.set_ylabel('Country')
axes.legend(loc = (1,0.9))
plt.show()
```
