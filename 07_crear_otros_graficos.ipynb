```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

# Histograma

bins sirve para ver el numero de barras que queremos ver para la distribucion de frecuencias

```python
plt.hist(data, bins=10, histtype='bar')
plt.show()
```

```python
plt.hist(data, bins=10, histtype='step')
plt.show()
```

# Boxplot

vert= False pone en horizontal el boxplot,
patch_artist=True sirve para poner los colores de las cajas
notch hace un enfoque hacia la mediana
showfliers sirve para mostrar los outliers es decir datos que se encuentran fuera de nuestra medicion


```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True )
plt.show()
```

```python
data = np.append(data,200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False )
plt.show()
```
