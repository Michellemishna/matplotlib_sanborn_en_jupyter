```python
import seaborn as sns
import matplotlib.pyplot as plt
```

displot es el grafico por excelencia, todos pueden ser modificados desde sus atributos, como kind con el que eliges que tipo de grafico sera

```python
tips = sns.load_dataset('tips')
tips
```

stat le da la indicacion de como quiero que trabaje con la informacion es decir por frecuencia, conteo, probabilidad, porcentaje, etc. 'fecuency', 'count' 'probability' 'percent', 'density' por default usa count
multiple sirve paraa decidir como presentara la info 'layer', 'dodge', 'stack', 'fill', usa 'layer como default

```python
sns.histplot(data=tips, x='tip', bins=15,cumulative=False, hue='sex', stat='count', multiple='dodge')
plt.show()
```

diagrama de densidad
cumulatif funciona para mostrar un acumulado de la informacion
fill hace un sombreado

```python
sns.kdeplot(tips,x='tip', hue='sex', fill=True, bw_adjust=1)
plt.show()
```

```python
sns.kdeplot(tips,x='tip', hue='sex', cumulative=True)
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
```

```python
sns.displot(data=tips, x='tip', hue='sex',kind='hist', multiple='stack')
```
